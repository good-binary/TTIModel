```python
from torch import autocast
from diffusers import StableDiffusionPipeline
import os


pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
pipe = pipe.to("cuda")


prompt = "a bowl of noodles in a vintage style"
with autocast("cuda"):
    image = pipe(prompt).images[0]

output_dir = "./generated_images"
os.makedirs(output_dir, exist_ok=True)
image_path = os.path.join(output_dir, "image.png")
image.save(image_path)
print(f"Image saved to {image_path}")
```

```python
from time import time
prompts = ["a bowl of noodles in a vintage style", "a futuristic cityscape at night", "a serene beach at sunset"]
for prompt in prompts:
    with autocast("cuda" if torch.cuda.is_available() else "cpu"):
        image = pipe(prompt).images[0]
    filename = f"{prompt.replace(' ', '_')}_{int(time())}.png"
    image_path = os.path.join(output_dir, filename)
    image.save(image_path)
    print(f"Image saved to {image_path}")
```

```python
from PIL import Image
# List of prompts
prompts = ["skincare products", "vitamin c serum", "lipbalm"]

images = []
for prompt in prompts:
    with autocast("cuda" if torch.cuda.is_available() else "cpu"):
        image = pipe(prompt).images[0]
        images.append(image)


grid_size = (1, len(prompts))
image_width, image_height = images[0].size
grid_image = Image.new("RGB", (grid_size[1] * image_width, grid_size[0] * image_height))

for i, image in enumerate(images):
    grid_image.paste(image, (i * image_width, 0))


output_dir = "./generated_images"
os.makedirs(output_dir, exist_ok=True)
grid_image_path = os.path.join(output_dir, f"image_grid_{int(time())}.png")
grid_image.save(grid_image_path)
print(f"Image grid saved to {grid_image_path}")
```
